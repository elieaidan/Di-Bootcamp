```python
def strip_dollar(s) -> str:
    new_s = s.strip('$')
    return float(new_s)
```

```python
df['item_price'] = df['item_price'].apply(strip_dollar)
```

```python
df['item_name'].value_counts()
```

```python
# Step 9. Which was the most ordered item?
df.groupby('item_name').sum(numeric_only=True).sort_values('quantity', ascending = False).head(1)
```

```python
# Step 10. How many items were ordered?
df['quantity'].sum()
```

```python
# Step 11. What was the most ordered item in the choice_description column?
df.groupby('choice_description').sum(numeric_only=True).sort_values('quantity', ascending = False).head(1)
```
