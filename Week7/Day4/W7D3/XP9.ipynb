Create a function called scatter_plot that takes the dataframe df and plots the relationship between the following two
features:
- displacement over x-axis
- acceleration over y-axis



```python
plt.scatter(df_mpg['displacement'], df_mpg['acceleration'])
plt.xlabel('displacement')
plt.ylabel('acceleration')
```

Create a function bar_plot that takes the dataframe df and compares the cylinders of all the cars from 1978 model_year and toyota company

```python
toyota = df_mpg[(df_mpg['car_name'].str.contains('toyota')) & (df_mpg['model_year'] > 78)]
toyota.head()
```

```python
non_toyota = df_mpg[(~df_mpg['car_name'].str.contains('toyota')) & (df_mpg['model_year'] > 78)]
```

```python
non_toyota
```

```python
plt.bar(non_toyota['model_year'], non_toyota['cylinders'], label = 'non toyota')
plt.bar(toyota['model_year'], toyota['cylinders'], label = 'toyota')
plt.legend()
```
