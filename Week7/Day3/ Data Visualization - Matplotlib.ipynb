```python
import numpy as np
import matplotlib.pyplot as plt 
```

```python
rand_int = np.random.randint(low=0, high=75, size=100)
```

```python
rand_int
```

```python
rand_int = rand_int.reshape((50,2))
```

```python
first_column = rand_int[:, 0] 
second_column = rand_int[:, 1]  # extract the second column
```

```python
plt.scatter(first_column, second_column)
plt.xlabel(first_column)
plt.ylabel(second_column)
```

```python
fig, ax=plt.subplots(1,1,figsize=(4, 4))
```

```python
ax.hist(first_column, color='g')
ax.hist(second_column, color='r')
```
